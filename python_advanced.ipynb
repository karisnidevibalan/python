```python
def averager():
    total = 0
    count = 0

    def add(n):
        nonlocal total, count
        total += n
        count += 1
        return total / count

    return add

avg = averager()

print(avg(1))  # 10.0
print(avg(20))  # 15.0
print(avg(30))  # 20.0
```

```python
from collections import deque

line = deque()

def join_line(name):
    line.append(name)
    print(name, "joined the line.")

def serve():
    if not line:
        print("No one is in the line.")
        return
    person = line.popleft()
    print("Serving:", person)

join_line("Rahul")
join_line("Ananya")
join_line("Kiran")

serve()
serve()

print("People still in line:", list(line))
```

```python
import os
NOTES_DIR="notes"
if not os.path.exists(NOTES_DIR):
  os.makedirs(NOTES_DIR)
  print(f"using {NOTES_DIR}")
else:
  print(f"working in current directory")
print(f"cwd:",os.getcwd())
def new_note(filename,text):
  path=os.path.join(NOTES_DIR,filename)
  with open(path,"w")as f:
    f.write(text+"\n")
  print(f"new note added to {path}")
new_note("example.txt", "i am in begginer stage")

def append_notes(filename,text):
  path=os.path.join(NOTES_DIR,filename)
  with open(path,"a")as f:
    f.write(text+"\n")
  print(f"added to {path}")
append_notes("example.txt","I am improving")
append_notes("example.txt","Doing well")
def read_note(filename):
  path=os.path.join(NOTES_DIR,filename)
  try:
    with open(path,"r") as f:
      fd=f.read()
      print(filename)
      print(fd)
      print("Content:\n", fd)
  except FileNotFoundError:
    print("file not found")
read_note("example.txt")
```

```python
def list_notes():
  files=os.listdir(NOTES_DIR)
  text_files=[f for f in files if f.endswith(".txt")]
  if not text_files:
    print("there are no notes available")
  else:
    for f in sorted(text_files) :
      print("-",f)
list_notes()
```

```python
def traffic_light_system(time_24hr, traffic_present, emergency=False):
    """
    time_24hr -> int (0 to 23)
    traffic_present -> True / False
    emergency -> True / False
    """


    if emergency:
        print("üö® Emergency vehicle detected!")
        print("üîµ Traffic Light: BLUE (Give full way)")
        return

    if 22 <= time_24hr or time_24hr < 5:
        print("üåô Night time detected")
        print("üüß Traffic Light: ORANGE (Blinking)")
        return

    if 5 <= time_24hr < 10 and not traffic_present:
        print("üåÖ Morning - No traffic detected")
        print("üüß Traffic Light: ORANGE (Blinking)")
        return

    print("üö¶ Normal Traffic Mode Activated")

    print("üî¥ RED - Stop")
    print("üü¢ GREEN - Go")
    print("üü° YELLOW - Slow Down / Prepare to Stop")

traffic_light_system(6, traffic_present=False)

print("\n---\n")

traffic_light_system(15, traffic_present=True, emergency=True)

print("\n---\n")

traffic_light_system(21, traffic_present=True)
```
