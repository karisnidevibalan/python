```python
import pandas as pd
data = pd.read_csv("/content/StudentsPerformance.csv")
print(data)
```

```python
print(f"Shape of the DataFrame: {data.shape}")
data.info()
```

```python
print('Missing values per column:')
print(data.isnull().sum())
```

```python
print(f"Shape of the DataFrame before handling duplicates: {data.shape}")
duplicate_rows = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")

if duplicate_rows > 0:
    data.drop_duplicates(inplace=True)
    print(f"Shape of the DataFrame after removing duplicates: {data.shape}")
    print("Duplicate rows removed successfully.")
else:
    print("No duplicate rows found.")
```

```python
numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns
print(f"Numerical columns identified: {list(numerical_cols)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 5))
for i, col in enumerate(numerical_cols):
    plt.subplot(1, len(numerical_cols), i + 1)
    sns.boxplot(y=data[col])
    plt.title(f'Box plot of {col}')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```

```python
print('Descriptive statistics for numerical columns:')
print(data[numerical_cols].describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 5))
for i, col in enumerate(numerical_cols):
    plt.subplot(1, len(numerical_cols), i + 1)
    sns.histplot(data[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

correlation_matrix = data[numerical_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Scores')
plt.show()
```

```python
categorical_cols = data.select_dtypes(include='object').columns
print(f"Categorical columns identified: {list(categorical_cols)}")
```

```python
print('Group-by analysis on numerical scores by categorical variables:')
for col in categorical_cols:
    print(f'\nMean scores grouped by {col}:')
    print(data.groupby(col)[numerical_cols].mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for cat_col in categorical_cols:
    plt.figure(figsize=(18, 6))
    plt.suptitle(f'Distribution of Numerical Scores by {cat_col}', fontsize=16)

    for i, num_col in enumerate(numerical_cols):
        plt.subplot(1, len(numerical_cols), i + 1)
        sns.boxplot(x=cat_col, y=num_col, data=data)
        plt.title(f'{num_col} by {cat_col}')
        plt.xlabel(cat_col)
        plt.ylabel(num_col)
        plt.xticks(rotation=45, ha='right')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Prepare data for plotting
gender_scores = data.groupby('gender')[numerical_cols].mean().reset_index()
prep_course_scores = data.groupby('test preparation course')[numerical_cols].mean().reset_index()
lunch_scores = data.groupby('lunch')[numerical_cols].mean().reset_index()

# Melt data for grouped bar charts
gender_scores_melted = gender_scores.melt(id_vars='gender', var_name='Score Type', value_name='Mean Score')
prep_course_scores_melted = prep_course_scores.melt(id_vars='test preparation course', var_name='Score Type', value_name='Mean Score')
lunch_scores_melted = lunch_scores.melt(id_vars='lunch', var_name='Score Type', value_name='Mean Score')

# Create a figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(20, 6), sharey=True)
fig.suptitle('Impact of Categorical Variables on Student Scores', fontsize=18)

# Plot Mean Scores by Gender
sns.barplot(x='gender', y='Mean Score', hue='Score Type', data=gender_scores_melted, ax=axes[0])
axes[0].set_title('Mean Scores by Gender')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Mean Score')
axes[0].legend(title='Score Type')

# Plot Mean Scores by Test Preparation Course
sns.barplot(x='test preparation course', y='Mean Score', hue='Score Type', data=prep_course_scores_melted, ax=axes[1])
axes[1].set_title('Mean Scores by Test Preparation Course')
axes[1].set_xlabel('Test Preparation Course')
axes[1].set_ylabel('') # y-label shared with first plot
axes[1].legend(title='Score Type')

# Plot Mean Scores by Lunch Type
sns.barplot(x='lunch', y='Mean Score', hue='Score Type', data=lunch_scores_melted, ax=axes[2])
axes[2].set_title('Mean Scores by Lunch Type')
axes[2].set_xlabel('Lunch Type')
axes[2].set_ylabel('') # y-label shared with first plot
axes[2].legend(title='Score Type')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
